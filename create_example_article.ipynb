Mettre à jour manifest.json pour ajouter une nouvelle catégorie « Examples » contenant l'article avec la date du jour au format Month DD, YYYY.

````python
from pathlib import Path
import json

article_path = Path('articles/example-category/example-article.md')
article_path.parent.mkdir(parents=True, exist_ok=True)

article_content = '''# Example Article

Welcome to the CipherOps example article. This page demonstrates the blog features, including math, code, graphs, and embedded PDFs.

## Inline Math and Display Equation
Here is inline math using $e^{i\pi} + 1 = 0$ inside a sentence.

And here is the famous Euler identity displayed below:

$$
e^{i\pi} + 1 = 0
$$

## Python Code Example
```python
import math

def compute_area(radius):
    return math.pi * radius ** 2

print(compute_area(5))
```

## Graph Definition
<div class="graph-definition">
{"nodes":[{"id":"A","label":"Node A","link":"https://example.com"},{"id":"B","label":"Node B","link":"articles/example-category/example-article.md"}],"edges":[{"from":"A","to":"B","directed":true}]}
</div>

## Embedded PDF
<div class="pdf-embed" data-src="pdfs/example.pdf"></div>

## Closing Thoughts
This article is a demonstration of CipherOps features. The graph and PDF are rendered dynamically, math is typeset, and code receives syntax highlighting.
'''

article_path.write_text(article_content, encoding='utf-8')

manifest_path = Path('manifest.json')
manifest = []
if manifest_path.exists():
    manifest = json.loads(manifest_path.read_text(encoding='utf-8'))

examples_entry = {
    'category': 'Examples',
    'articles': [
        {
            'title': 'Example Article',
            'file': 'articles/example-category/example-article.md',
            'date': 'June 1, 2026'
        }
    ]
}

existing = next((item for item in manifest if item.get('category') == 'Examples'), None)
if existing is None:
    manifest.append(examples_entry)
else:
    existing['articles'] = [examples_entry['articles'][0]]

manifest_path.write_text(json.dumps(manifest, indent=2), encoding='utf-8')
````
